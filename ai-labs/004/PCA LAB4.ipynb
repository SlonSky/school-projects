### Функция отрисовки классов (универсальна по измерениям)

```python
def draw_classes(classes, ax, msize=9):
    marker_colors = [
        (1, 0, 0),
        (0, 1, 0),
        (0, 0, 1),
        (1, 0, 1),
        (1, 1, 1),
    ]

    line_markers = 'vspPp*'
    
    cls = lambda cl_num: [classes[cl_num, :, i] for i in range(classes.shape[2])]
    
    for i in range(len(classes)):
        plt.plot(*cls(i), line_markers[i], color=marker_colors[i], markersize=msize)
    
    ax.set_xlabel('x1')
    
    if classes.shape[2] > 1:
        ax.set_ylabel('x2')
        
    if classes.shape[2] > 2:
        ax.set_zlabel('x3')
    
    ax.legend(['Class %d' % i for i in range(len(classes))])
```

## 3. Реалізувати програму для виділення основних компонентів на основі кореляційногоф підходу для заданих векторів-образів.

### Функция вычисления средней автокорелляции по объектам в классе

```python
def autocorr_per_obj(cls):
    # reshape to vertical vectors
    vert_objs = cls.reshape(cls.shape[0], -1, 1)
    autocorr = [i.dot(i.T) for i in vert_objs]
    mean_autocorr = np.mean(autocorr, axis=0)
    
    return mean_autocorr
```

### Функция вычисления средней автокорелляции по классам

```python
def autocorr_per_clss(clss):
    autocorr = [autocorr_per_obj(i) for i in clss]
    return np.mean(autocorr, axis=0)
```

### Функция нахождения матрицы проекции

```python
def proj_mat(R, dim):
    return np.linalg.eig(R)[1][:, :dim].T
```

### Функция полного алгоритма метода главных компонент

```python
def pca(classes, dim):
    R = autocorr_per_clss(classes)
    S = proj_mat(R, dim)
    out = list()
    for i in classes:
        out.append([S.dot(j) for j in i])
    
    return np.array(out)
```
