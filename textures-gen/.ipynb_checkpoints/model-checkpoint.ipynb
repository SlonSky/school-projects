```python
import os
import PIL.Image as Image

import numpy as np
import matplotlib.pyplot as plt

from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

## Load data

```python
W = 128
H = 128

def load_imgs(dirname):
    return np.array([
        np.asarray(Image.open(os.path.join(dirname, i))
        .resize((H, W)))
        for i in os.listdir(dirname)
    ])
    
```

```python
metal = load_imgs('metal')
marble = load_imgs('marble')
wood = load_imgs('wood')
stone = load_imgs('stone')
```

```python
metal.shape
```

```python
marble.shape
```

```python
wood.shape
```

```python
stone.shape
```

```python
X = np.concatenate([metal, marble, wood, stone])
```

```python
X.shape
```

```python
METAL = 0
MARBLE = 1
WOOD = 2
STONE = 3

Y = np.concatenate([
    np.full((metal.shape[0]), METAL),
    np.full((marble.shape[0]), MARBLE),
    np.full((wood.shape[0]), WOOD),
    np.full((stone.shape[0]), STONE),
])
```

```python
Y = to_categorical(Y).astype(np.float32)
```

```python
Y.shape
```

## Prepare data

```python
def normalize(batch):
    return batch.astype('float32') / 255.

def gray(batch):
    return np.mean(batch, axis=3).reshape(-1, H, W, 1)
```

```python
TEST_SIZE = 0.2
CH = 3

X_prepared = normalize(X)

if CH == 1:
    X_prepared = gray(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_prepared, Y, test_size=TEST_SIZE,
                                                   random_state=666)
```

## Making models

```python
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model

def create_deep_conv_ae():
    input_img = Input(shape=(H, W, CH))

    x = Conv2D(64, (8, 8), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)
    
    x = Conv2D(32, (4, 4), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)
    
    x = Conv2D(16, (4, 4), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2), padding='same')(x)
    
    encoded = Conv2D(3, (16, 16), activation='relu', padding='same')(x)

    input_encoded = Input(shape=(32, 32, 3))
    
    x = Conv2D(16, (4, 4), activation='relu', padding='same')(input_encoded)
    x = UpSampling2D((2, 2))(x)
    
    x = Conv2D(32, (8, 8), activation='relu', padding='same')(x)
    x = UpSampling2D((2, 2))(x)
    
    decoded = Conv2D(3, (16, 16), activation='sigmoid', padding='same')(x)

    # Модели
    encoder = Model(input_img, encoded, name="encoder")
    decoder = Model(input_encoded, decoded, name="decoder")
    autoencoder = Model(input_img, decoder(encoder(input_img)), name="autoencoder")
    return encoder, decoder, autoencoder

c_encoder, c_decoder, c_autoencoder = create_deep_conv_ae()
c_autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

c_autoencoder.summary()

```

```python
from keras.models import load_model

c_encoder, c_decoder, c_autoencoder = load_model('encoder-v2-'), load_model('decoder-v2-'), load_model('autoencoder-v2-')
```

### Generate textures based of LERP from one material image in latent space to another (16 steps)

```python
plt.figure(figsize=(7,14))
plt.subplot(1, 2, 1)
plt.imshow(X_test[4])
plt.title('FROM')

plt.subplot(1, 2, 2)
plt.imshow(X_test[12])
plt.title('TO')

plt.figure(figsize=(14,14))

# stone
encoded_start = (c_encoder.predict(X_test[4].reshape(1, 128, 128, 3)))
encoded_end = (c_encoder.predict(X_test[12].reshape(1, 128, 128, 3)))

for i in range(1, 17):
    res = c_decoder.predict((16 - i - 1)/16 * encoded_start + (i - 1)/16*encoded_end)
    plt.subplot(4,4,i)
    plt.imshow(res[0])
    plt.title(str(i))
        
```

### A kind of style transfering (mix of two images from different materials)

```python

plt.figure(figsize=(10, 10))
plt.subplot(3,2,1)

first = X_test[3]
plt.imshow(first)
plt.title('First')

plt.subplot(3,2,2)

second = X_test[4]
plt.imshow(second)
plt.title('Second')

encoded1 = c_encoder.predict(first.reshape(1, 128, 128, 3))
                            
plt.subplot(3,2,3)
plt.imshow(encoded1[0])
plt.title('First encoded')

encoded2 = c_encoder.predict(second.reshape(1, 128, 128, 3))
                            
plt.subplot(3,2,4)
plt.imshow(encoded2[0])
plt.title('Second encoded')

         
plt.subplot(3,2,5)
enc_res = (0.6 * encoded1 +  0.4 * encoded2)

plt.imshow(enc_res[0])
plt.title('Mix')

plt.subplot(3,2,6)
res = c_decoder.predict(enc_res)
plt.imshow(res[0])
plt.title('Result');
```
