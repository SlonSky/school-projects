```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -20
b = 20
width = 30

np.random.seed(666)

X = np.random.uniform(a, b, size=(width,))
Y = np.random.uniform(a, b, size=(width,))

plt.plot(X, Y, '-g') 
plt.show()
```

```python
def bubble_sort(A, order=0):
    l = A.shape[0]
    for i in range(l):
        for j in range(l):
            if (A[i] < A[j]) and order == 0:
                A[i], A[j] = A[j], A[i]
                
            if (A[i] > A[j]) and order == 1:
                A[i], A[j] = A[j], A[i]
    return A

X = bubble_sort(X)

plt.plot(X, Y, '.r')
plt.show()
```
